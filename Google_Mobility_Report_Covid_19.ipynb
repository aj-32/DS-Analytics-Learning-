```python
# importing all the required packages

import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

Pulling the data directly from the webpage
Since we are pulling the data directly using the URL, we'll get fresh data every time we run it

```python
df = pd.read_csv('https://www.gstatic.com/covid19/mobility/Global_Mobility_Report.csv')
df.head()
```

Each column represents each category so it's very important to understand who's doing what, when and where.

```python
df.columns
```

In this dataset, the columns 'country_region' and 'sub_region' store the values for Country and Provinces / State

```python
# using the .loc function to filter data based on certain conditions and obtain the required data 

NY_df = df.loc[(df['country_region']=='United States')&
               (df['sub_region_1']=='New York')]
```

```python
NY_df.head(10)
```

converting the column 'sub_region_2' into a List

```python
list(NY_df['sub_region_2'])
```

Removing a lot of duplicate values

```python
set(list(NY_df['sub_region_2']))

# set function returns all the unique values
```

```python
M_df = NY_df.loc[NY_df['sub_region_2']=='New York County']
```

```python
M_df.head()
```

Descriptive Analysis

```python
# finding the max value of retail and recreation for Manhattan region using the max function

M_df['retail_and_recreation_percent_change_from_baseline'].max()

```

```python
# On which dates the value of 'retail_and_recreation_percent_change_from_baseline' was equal to the max value i.e. 6.0

M_df['date'].loc[M_df['retail_and_recreation_percent_change_from_baseline']==6.0]
```

```python
# finding the minimum value of retail and recreation for Manhattan region using the min function

M_df['retail_and_recreation_percent_change_from_baseline'].min()
```

```python
M_df['date'].loc[M_df['retail_and_recreation_percent_change_from_baseline']==-89.0]

# here we can see that the least movement happened in April 2020 and July 2021 when Covid 19 was rampant in NY during this time period
```

```python
M_df['retail_and_recreation_percent_change_from_baseline'].mean()

## with a mean of negative 52 the data indicates how Covid 19  
```

Visualizing the Mobility trend

```python
# Plotting trends to see the big picture 

M_df['date']=pd.to_datetime(M_df['date'])
```

```python
M_df = M_df.set_index('date')
```

```python
M_df.head()

# The date column now becomes the index of the dataset and this was done by using the 'to_datetime function' and the 'set_index' function
```

```python
# plotting the movement trend for retail and recreation

M_df['retail_and_recreation_percent_change_from_baseline'].plot()
plt.show()

# We can see that since the lockdown in March, the movement has dramatically decreased and 
# then it is following a wayward line depending on the intensity of the cases that is being recorded
```
